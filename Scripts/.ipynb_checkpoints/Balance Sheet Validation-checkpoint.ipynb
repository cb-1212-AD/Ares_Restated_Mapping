```python
hurtigruten_cv_df = pd.read_excel("C:\\Users\curt.beck\\Downloads\\Hurtigruten.xlsx", sheet_name="Balance_Sheet", skiprows=6)
```

```python
item_list = hurtigruten_cv_df[hurtigruten_cv_df.columns[0]].tolist()
end_match = hurtigruten_cv_df[hurtigruten_cv_df.columns[0]].str.extract(r'(Total\s+Liabilities\s+and\s+Equity)').dropna().values[0][0]
end_index = item_list.index(end_match) + 1
hurtigruten_cv_df.set_index(hurtigruten_cv_df.columns[0], inplace=True)
filtered_df = hurtigruten_cv_df.iloc[0: end_index]
clean_df = filtered_df.dropna(axis=1, thresh=1)
clean_final_df = clean_df.dropna(thresh=1)
clean_final_df[[clean_final_df.columns]] = clean_final_df[[clean_final_df.columns]].apply(pd.to_datetime, axis=1)
```
